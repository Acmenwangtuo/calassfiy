```python
import os
import shutil
import glob
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
label_path = '/home/bnc/tool/HistomicsML/yourproject/classify/label/'
image_path = '/home/bnc/tool/HistomicsML/yourproject/classify/images/'
crop_path = '/home/bnc/tool/HistomicsML/yourproject/classify/crop/'
```

```python
def get_image(label):
    mask = cv2.imread(label,1)
    gray = cv2.cvtColor(mask,cv2.COLOR_BGR2GRAY)
    print(gray.shape)
    gray[gray == 255] = 1
#     plt.imshow(gray)
    print(np.unique(gray))
#     plt.show()
    return gray
def crop(label,gray,image_path,crop_path,step=10):
    h,w = gray.shape
    label_name = os.path.basename(label).split('_')[0]
    image = cv2.imread(os.path.join(image_path,label_name+'.png'))
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    for i in range(0,h,step):
        for j in range(0,w,step):
            p = i if i+50 <= h else h-50
            q = j if j+50 <= w else w-50
            temp = gray[p:p+50,q:q+50]
            if np.sum(temp==1) == 2500:
                rgb = image[p:p+50,q:q+50,::]
                cv2.imwrite(crop_path+label_name+'_{}_{}.png'.format(p,q),rgb)
            else:
                continue
```

```python
def main():
    list_label = glob.glob(label_path+'*.png')
    for label in list_label:
        gray = get_image(label)
        crop(label,gray,image_path,crop_path)
main()
```
